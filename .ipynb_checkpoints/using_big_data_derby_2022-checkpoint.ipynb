```python
race_table_sorted = race_table.sort_values(by=['race_date', 'race_number', 'program_number'], ascending=True)
```

```python
race_table_sorted = race_table_sorted.reset_index(drop=True)
```

```python
race_table_sorted
```

```python
race_table_d = pd.get_dummies(race_table_sorted)
```

```python
race_table_d
```

```python
race_table_d_unique = race_table_d['race_date'].unique()
```

```python
race_table_d_unique
```

```python
print(len(race_table_d_unique))
train_len = round(len(race_table_d_unique)*0.7)
print(train_len)
last_race_date_train = race_table_d_unique[train_len]
print(last_race_date_train)

tmp_idx = race_table_d[race_table_d['race_date'] == last_race_date_train].index.values
print(tmp_idx)
last_race_index_train = tmp_idx[-1]
print(last_race_index_train)
```

```python
train = race_table_d.iloc[0:last_race_index_train+1]
```

```python
test = race_table_d.iloc[last_race_index_train+1:]
```

```python
print(len(train))
print(len(test))
print(len(race_table_d))
print(len(train)/len(race_table_d))
print(len(test)/len(race_table_d))
```

```python
#%pip install imbalanced-learn
```

```python
train = train.set_index('race_date')
test = test.set_index('race_date')
```

```python
train['finish_pos'].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler

rank_1 = train['finish_pos'].value_counts()[1]
rank_2 = train['finish_pos'].value_counts()[2]
rank_3 = train['finish_pos'].value_counts()[3]
# make rank_4 = rank_1

#print(rank_1, rank_2, rank_3, rank_4)
print(rank_1, rank_2, rank_3)

rus = RandomUnderSampler(sampling_strategy={1:rank_1,2:rank_2,3:rank_3,4:rank_1}, random_state=0)
```

```python
X_train = train.drop(['finish_pos'], axis=1)
y_train = train['finish_pos']
X_test = test.drop(['finish_pos'], axis=1)
y_test = test['finish_pos']
```

```python
X_train_rus, y_train_rus = rus.fit_resample(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=41)
```

```python
clf.fit(X_train_rus, y_train_rus)
print(clf.score(X_train,y_train), clf.score(X_test,y_test))
```
